```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# The NumPy.linspace() function returns an array of evenly spaced values within 
# the specified interval [start, stop].

def simulate_geometric_random_walk(S0, T=2, N=1000, mu=0.1, sigma=0.05):
    dt = T/N
    t = np.linspace(0, T, N)
    # standard normal distribution N(0, 1)
    W = np.random.standard_normal(size=N)
    # N(0, dt) = np.cumsum(W) * np.sqrt(dt)
    W = np.cumsum(W) * np.sqrt(dt)

    X = (mu - 0.5 * sigma ** 2) * t + sigma * W
    S = S0 * np.exp(X)

    return t, S
```

```python
def plot_simulation(t, S):
    plt.plot(t, S)
    plt.xlabel('Time (t)')
    plt.ylabel('Stock Price S(t)')
    plt.title('Geometric brownian motion')
    plt.show()
```

```python
t, S = simulate_geometric_random_walk(10)
plot_simulation(t, S)
```
